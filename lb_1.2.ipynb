```python
# LAB 1.2: Python Data Structures

# SCENARIO I: Managing a List of Sensor Readings
sensor_readings = [23.5, 25.2, 26.8, 24.1, 27.3]

# Retrieve first, last, and middle elements
first_reading = sensor_readings[0] if sensor_readings else None
last_reading = sensor_readings[-1] if sensor_readings else None
mid_index = len(sensor_readings) // 2
middle_reading = sensor_readings[mid_index] if sensor_readings else None

# Filter readings above a threshold
threshold = 25.0
high_readings = [sensor_readings[i] for i in range(len(sensor_readings)) if sensor_readings[i] > threshold]

# Calculate average of non-zero readings
average = sum(sensor_readings) / len(sensor_readings) if sensor_readings else 0

# Display results
print("Sensor Readings Analysis:")
print(f"First Reading: {first_reading}")
print(f"Last Reading: {last_reading}")
print(f"Middle Reading: {middle_reading}")
print(f"Readings > 25°C: {high_readings}")
print(f"Average Reading: {average:.2f}")

# SCENARIO II: Managing a Dictionary of Customer Data
customers = {
    "C001": [50.0, 75.0, 25.0],
    "C002": [120.0, 80.0],
    "C003": [30.0]
}

# Retrieve purchases for a specific customer (“C001”)
customer_id = "C001"
purchases = customers.get(customer_id, [])

# Calculate total purchases for a customer
total_purchases = sum(purchases) if purchases else 0

# Find customers with purchases above a threshold
threshold = 100
high_spenders = {cid: sum(p) for cid, p in customers.items() if sum(p) > threshold}

# Display results
print("\nCustomer Data Analysis:")
print(f"Purchases for C001: {purchases}")
print(f"Total for C001: {total_purchases}")
print(f"High Spenders (>100): {high_spenders}")

# SCENARIO III: Using Tuples for Immutable Data Points
coordinates = [(1.0, 2.0), (-1.0, 3.0), (4.0, 5.0), (2.0, -1.0)]

# Retrieve first point’s x-coordinate
first_x = coordinates[0][0] if coordinates else None

# Extract all y-coordinates
y_coords = [point[1] for point in coordinates]

# Find points in the first quadrant (x > 0, y > 0)
first_quadrant = [point for point in coordinates if point[0] > 0 and point[1] > 0]

# Display results
print("\nCoordinates Analysis:")
print(f"First point's x-coordinate: {first_x}")
print(f"All y-coordinates: {y_coords}")
print(f"First quadrant points: {first_quadrant}")

# SCENARIO IV: Using Sets for Unique Categories
categories = ["Electronics", "Clothing", "Electronics", "Books", "Toys"]

# Convert list to set to get unique categories
unique_categories = set(categories)

# Check if a specific category exists
target_category = "Electronics"
has_electronics = target_category in unique_categories

# Find common categories with another dataset
other_categories = {"Clothing", "Books", "Electronics"}
common_categories = unique_categories.intersection(other_categories)

# Display results
print("\nCategories Analysis:")
print(f"Unique Categories: {unique_categories}")
print(f"Has Electronics: {has_electronics}")
print(f"Common Categories: {common_categories}")
```
