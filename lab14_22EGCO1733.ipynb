```python
import numpy as np

# STEP 1: Create 1D array of temperature readings (float)
temps = [23.5, 25.2, 26.8, 24.1, 27.3]
temps_array = np.array(temps)

# STEP 2: Operations on temps_array
temps_shape = temps_array.shape
temps_ndim = temps_array.ndim
temps_size = temps_array.size
temps_dtype = temps_array.dtype
temps_slice = temps_array[:3]

# STEP 3: Arithmetic Operations
temps_plus_2 = temps_array + 2
temps_minus_1 = temps_array - 1
temps_squared = temps_array ** 2
temps_sqrt = np.sqrt(temps_array)

# STEP 4: Statistical operations
temps_sum = np.sum(temps_array)
temps_mean = np.mean(temps_array)
temps_std = np.std(temps_array)
temps_max = np.max(temps_array)
temps_min = np.min(temps_array)
temps_median = np.median(temps_array)

# STEP 5: Other Operations
temps_reshaped = temps_array.reshape(5, 1)
temps_unique = np.unique(temps_array)
temps_sorted = np.sort(temps_array)
temps_argsort = np.argsort(temps_array)
temps_where = np.where(temps_array > 25)

# STEP 6: Display results for temps_array
print("Temperature Array Operations:")
print(f"Shape: {temps_shape}")
print(f"Number of Dimensions: {temps_ndim}")
print(f"Size: {temps_size}")
print(f"Data Type: {temps_dtype}")
print(f"Slice (first 3 elements): {temps_slice}")
print(f"Add 2: {temps_plus_2}")
print(f"Subtract 1: {temps_minus_1}")
print(f"Squared: {temps_squared}")
print(f"Square Root: {temps_sqrt}")
print(f"Sum: {temps_sum:.2f}")
print(f"Mean: {temps_mean:.2f}")
print(f"Standard Deviation: {temps_std:.2f}")
print(f"Max: {temps_max}")
print(f"Min: {temps_min}")
print(f"Median: {temps_median}")
print(f"Reshaped (5x1):\n {temps_reshaped}")
print(f"Unique Values: {temps_unique}")
print(f"Sorted: {temps_sorted}")
print(f"Argsort (indices): {temps_argsort}")
print(f"Where (indices > 25): {temps_where}")
```
